```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('Automobile_data.csv')
print(df.info())
```

```python
df.head()
```

```python
# 1. what is the average price and average mileage for convertible cars
```

```python
cond1 = df['body-style'].str.lower()=='convertible'
df_covert=df[cond1]
df_covert
```

```python
avg_price_covert=df_covert['price'].mean()
avg_mileage_covert=df_covert['average-mileage'].mean()
dfa1=pd.DataFrame([avg_price_covert,avg_mileage_covert],columns=['average'],index=['price_covert','mileage-covert']).reset_index()
dfa1
```

```python
#  2. Sort cars by company and price in decending order
```

```python
dfa2=df.sort_values(by=['company','price'],ascending=False)
dfa2
```

```python
# 3. Save data for 'Toyota' separately
```

```python
cond3=df['company'].str.lower()=='toyota'
df3a=df[cond3]
df3a.to_csv("toyota_data_only.csv",index=False)
```

```python
 #4. which company is producing most expensive car?
```

```python
max_price=df['price'].max()
df[df['price']==max_price][['company','price']]
```

```python
# 5. Name the cars on basis of maximum length, highest wheelbase, maximum horsepower
```

```python
df.columns
col_req=['length','wheel-base','horsepower']
df5a=pd.DataFrame()
for i in col_req:
     df_temp=df[df[i]==df[i].max()][[i,'company','body-style']]
     df_temp.rename(columns={i:'value'},inplace=True)
     df_temp.iloc[0,0]=i
     df5a=pd.concat([df_temp,df5a])
display(df5a)    
```

```python
# 6. company wise average price of cars and order them in decending order
```

```python
df6a=df.groupby(['company'],as_index=False)['price'].mean().sort_values(by=['price'],ascending=False)
df6a
```

```python
# 7. create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"
```

```python
cond1=df['wheel-base']<100
cond2=df['length']<175
df['decision']=np.where(cond1 & cond2,"don't buy","buy")
df
```

```python
 #8. Select only cars whose decision is "buy"; find average price based on company of car
```

```python
df8a=df[df['decision']=='buy']
df8a.groupby(['company'],as_index=False)['price'].mean()
```

```python
#how to save multiple data frames to excel in python
```

```python
writer = pd.ExcelWriter('auto_summary.xlsx',engine='xlsxwriter')   
df.to_excel(writer,sheet_name='complete data',index=False)   
dfa1.to_excel(writer,sheet_name='ans 1',index=False)
dfa2.to_excel(writer,sheet_name='ans 2',index=False)
df3a.to_excel(writer,sheet_name='ans 3',index=False)
writer.close()
```
